# Foreword

This script downloads GPM 2D 30-minute frequency sampling and 0.1 arcdeg horizontal resolution precipitation data from 3IMERGHH.06 dataset, given as inputs a polygon representing the Area of Interest (AOI) and start/end date.

Important note when preparing the URLs:
- Version V06B runs for the period 20000601-S000000 to 20220508-S153000.
- Version V06C runs for the period 20220508-S160000 to 20230701-S133000.
- Version V06D runs for the period 20230701-S140000 to 20231108-S013000.
- Version V06E runs for the period 20231108-S020000 to present.

# Functions

```python
# Generate global 0.1 arcdeg GPM grid
def generate_grid_gpm(): 
    # Define the range of longitude and latitude
    lon_range = np.arange(-180, 180.1, 0.1)
    lat_range = np.arange(-90, 90.1, 0.1)
    
    # Create meshgrid of longitude and latitude
    lon, lat = np.meshgrid(lon_range, lat_range)
    
    # Flatten the meshgrid
    lon_flat = lon.flatten()
    lat_flat = lat.flatten()
    
    # Create a DataFrame from the flattened longitude and latitude
    grid_df = pd.DataFrame({
        'longitude': lon_flat,
        'latitude': lat_flat,
        'minx': lon_flat,
        'miny': lat_flat,
        'maxx': lon_flat + 0.1,
        'maxy': lat_flat + 0.1
    })
    
    # Create longitude and latitude ascending index starting from 0
    grid_df['lon_index'] = [int(round(x)) for x in (grid_df['longitude'] + 180) / 0.1]
    grid_df['lat_index'] = [int(round(x)) for x in (grid_df['latitude'] + 90) / 0.1]
    
    # Round longitude and latitude to avoid precision issues
    grid_df['longitude'] = grid_df['longitude'].round(6)
    grid_df['latitude'] = grid_df['latitude'].round(6)
    
    return grid_df

# Associate index to each grid cell of GPM grid
def extract_index(grid_df, given_lat, given_lon):

    # Filter the DataFrame based on the given latitude and longitude
    row = grid_df[(grid_df['latitude'] <= given_lat) & (grid_df['longitude'] <= given_lon) & 
    (grid_df['latitude'] + 0.1 >= given_lat) & (grid_df['longitude'] + 0.1 >= given_lon)]

    return row

## Multi-satellite precipitation estimate with gauge calibration - Final Run (recommended for general use) (GPM_3IMERGHH v06): https://disc.gsfc.nasa.gov/information/glossary?keywords=giovanni%20measurements&title=Giovanni%20Measurement%20Definitions%3A%20Precipitation
## Multi-satellite precipitation estimate with climatological gauge calibration - Late Run (GPM_3IMERGHHL v06): https://disc.gsfc.nasa.gov/information/glossary?keywords=giovanni%20measurements&title=Giovanni%20Measurement%20Definitions%3A%20Precipitation
# Generate GPM url files of precipitation, given start/end date and the GPM grid cell index
def generate_urls(start_date, end_date, output_file, index_min, index_max):
    base_url = "https://gpm1.gesdisc.eosdis.nasa.gov/opendap/GPM_L3/GPM_3IMERGHHL.06/"
    
    current_date = start_date
    sequence = 0
    
    with open(output_file, 'w') as file:
        while current_date <= end_date:
            # Reset sequence to 0 at the start of each day
            if current_date.hour == 0 and current_date.minute == 0:
                sequence = 0
            
            start_time = current_date.strftime("%Y%m%d-S%H%M%S")
            end_time = (current_date + timedelta(minutes=30) - timedelta(seconds=1)).strftime("E%H%M%S")
            sequence_str = str(sequence).zfill(4)

            # Determine the version based on the start_time
            if start_time >= '20231108-S020000':
                version = 'V06E'
            elif start_time >= '20230701-S140000':
                version = 'V06D'
            elif start_time >= '20220508-S160000':
                version = 'V06C'
            else:
                version = 'V06B'
            
            url = (base_url + current_date.strftime("%Y/%j/") +
                   f"""3B-HHR-L.MS.MRG.3IMERG.{start_time}-{end_time}.{sequence_str}.{version}.HDF5.nc4?precipitationCal[0:0][{index_min['lon_index'][0]}:{index_max['lon_index'][0]}][{index_min['lat_index'][0]}:{index_max['lat_index'][0]}],time,lon[{index_min['lon_index'][0]}:{index_max['lon_index'][0]}],lat[{index_min['lat_index'][0]}:{index_max['lat_index'][0]}]""")
            file.write(url + '\n')
            
            current_date += timedelta(minutes=30)
            sequence += 30  # Increment by 30 at each iteration
            
            if current_date > end_date:  # Check if we reached the end date
                break


# Download URLs in parallel
def download_urls(urls, num_processes): 
    
    # Calculate the number of URLs per process
    urls_per_process = len(urls) // num_processes

    # Create processes
    processes = []
    for i in range(num_processes):
        
        start_idx = i * urls_per_process
        end_idx = start_idx + urls_per_process if i < num_processes - 1 else len(urls)
        process_urls = urls[start_idx:end_idx]

        # Create a command for each process
        cmd = f"""
        wget --load-cookies ~/.urs_cookies --save-cookies ~/.urs_cookies --keep-session-cookies -i - <<<"{os.linesep.join(process_urls)}"
        """
        # Start the process
        process = subprocess.Popen(cmd, shell=True)
        processes.append(process)

    # Wait for all processes to complete
    for process in processes:
        process.wait()

# List all the files in directory, including folders and subfolder (recursive)
def list_files_in_directory(directory):
    file_names = []
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            file_names.append(file_path)
    return file_names

# Delete repeated files (repeated files have the suffix '.1')
def delete_files_with_suffix(directory, suffix):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(suffix):
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"Deleted: {file_path}")
                except OSError as e:
                    print(f"Error: {file_path} : {e.strerror}")

# Count files in folder to make sure we have all the files per year
def count_files_in_folders(folder):
    folder_names = []
    file_counts = []
    
    for root, dirs, files in os.walk(folder):
        for dir_name in dirs:
            folder_path = os.path.join(root, dir_name)
            file_count = sum([len(files) for r, d, files in os.walk(folder_path)])
            folder_names.append(dir_name)
            file_counts.append(file_count)
    
    df = pd.DataFrame({'Folder Name': folder_names, 'Number of Files': file_counts})
    return df

# Concatenate .nc files per year
def process_and_save_files(year, folder, folder_year_files, chunk_size, spatial_ref_path):
    # Find all files for the given year
    files = sorted(glob.glob(f'{folder}/{year}/*]'))  # Assuming .nc files
    if not files:
        return  # Skip years with no files

    # Use dask to handle large datasets
    with ProgressBar():
        ds_gpm = xr.open_mfdataset(files, concat_dim='time', combine='nested', parallel=True, chunks=chunk_size)

    # Process dataset
    ds_gpm = ds_gpm.rename({'precipitationCal': 'precip'})
    ds_gpm = ds_gpm['precip'] * 0.5  # Convert precipitation intensity to 30min precipitation depths

    # Add spatial reference
    sr = xr.open_dataset(spatial_ref_path)['spatial_ref']
    ds_gpm.coords['spatial_ref'] = sr

    # Save to file with Dask
    fn = f'{folder_year_files}/GPM_3IMERGHHL.06_{year}.nc'
    with ProgressBar():
        ds_gpm.to_netcdf(fn, mode='w', compute=True)
    print(f"Processed and saved data for year {year}")

import glob
import xarray as xr
from dask.diagnostics import ProgressBar

def process_and_save_files(year, folder, folder_year_files, chunk_size, spatial_ref_path):
    # Find all files for the given year
    files = sorted(glob.glob(f'{folder}/{year}/*.nc'))  # Assuming .nc files
    if not files:
        return  # Skip years with no files

    # Use dask to handle large datasets
    with ProgressBar():
        ds_gpm = xr.open_mfdataset(files, concat_dim='time', combine='nested', parallel=True, chunks=chunk_size)

    # Get the total length of the time dimension
    total_time_length = ds_gpm.sizes['time']

    # Calculate the midpoint
    midpoint = total_time_length // 2

    # Split the dataset into two parts
    ds_gpm_part1 = ds_gpm.isel(time=slice(0, midpoint))
    ds_gpm_part2 = ds_gpm.isel(time=slice(midpoint, total_time_length))

    # Process dataset part 1
    ds_gpm_part1 = ds_gpm_part1.rename({'precipitationCal': 'precip'})
    ds_gpm_part1 = ds_gpm_part1['precip'] * 0.5  # Convert precipitation intensity to 30min precipitation depths

    # Process dataset part 2
    ds_gpm_part2 = ds_gpm_part2.rename({'precipitationCal': 'precip'})
    ds_gpm_part2 = ds_gpm_part2['precip'] * 0.5  # Convert precipitation intensity to 30min precipitation depths

    # Add spatial reference
    sr = xr.open_dataset(spatial_ref_path)['spatial_ref']
    ds_gpm_part1.coords['spatial_ref'] = sr
    ds_gpm_part2.coords['spatial_ref'] = sr

    # Save the first part to file with Dask
    fn_part1 = f'{folder_year_files}/GPM_3IMERGHHL.06_{year}_part1.nc'
    with ProgressBar():
        ds_gpm_part1.to_netcdf(fn_part1, mode='w', compute=True)
    print(f"Processed and saved data for year {year} - part 1")

    # Save the second part to file with Dask
    fn_part2 = f'{folder_year_files}/GPM_3IMERGHHL.06_{year}_part2.nc'
    with ProgressBar():
        ds_gpm_part2.to_netcdf(fn_part2, mode='w', compute=True)
    print(f"Processed and saved data for year {year} - part 2")
```

# Postprocessing

```python
import os
import glob
import xarray as xr
import pandas as pd
from dask.diagnostics import ProgressBar
import gc

def process_and_save_files(year, folder, folder_year_files, chunk_size, spatial_ref_path):
    # Find all files for the given year
    files = sorted(glob.glob(f'{folder}/{year}/*]'))  # Assuming .nc files
    if not files:
        print(f"No files found for year {year}.")
        return  # Skip years with no files
    
    # Split files into three parts
    total_files = len(files)
    third = total_files // 3
    files_part1 = files[:third]
    files_part2 = files[third:2*third]
    files_part3 = files[2*third:]

    # Function to process and save part of the dataset
    def process_and_save_part(files_part, part_number):
        fn_part = f'{folder_year_files}/GPM_3IMERGHHL.06_{year}_part{part_number}.nc'
        if os.path.exists(fn_part):
            print(f"File {fn_part} already exists, skipping.")
            return
        else:
            # Read data from NetCDF files into DataFrames
            try:
                dataframes = []
                for file in files_part:
                    ds = xr.open_dataset(file)
                    df = ds.to_dataframe().reset_index()
                    dataframes.append(df)
            except Exception as e:
                print(f"Error reading files part {part_number}: {e}")
                return
            
            # Concatenate DataFrames
            try:
                concatenated_df = pd.concat(dataframes)
                concatenated_df.set_index(['time', 'lat', 'lon'], inplace=True)
                print(f"Concatenated data for part {part_number}.")
            except Exception as e:
                print(f"Error concatenating data for part {part_number}: {e}")
                return
            
            # Convert the concatenated DataFrame to xarray Dataset
            try:
                ds_gpm = concatenated_df.to_xarray()
                ds_gpm = ds_gpm.rename({'precipitationCal': 'precip'})
                ds_gpm['precip'] = ds_gpm['precip'] * 0.5  # Convert precipitation intensity to 30min precipitation depths
                print(f"Converted DataFrame to Dataset for part {part_number}.")
            except Exception as e:
                print(f"Error converting DataFrame to Dataset for part {part_number}: {e}")
                return

            # Add spatial reference
            try:
                sr = xr.open_dataset(spatial_ref_path)['spatial_ref']
                ds_gpm.coords['spatial_ref'] = sr
            except Exception as e:
                print(f"Error adding spatial reference for part {part_number}: {e}")
                return

            # Save the Dataset to a NetCDF file
            try:
                with ProgressBar():
                    ds_gpm.to_netcdf(fn_part, mode='w', compute=True)
                print(f"Processed and saved data for year {year} - part {part_number}")
            except Exception as e:
                print(f"Error saving part {part_number} for year {year}: {e}")

            # Explicitly delete the dataset and call garbage collection
            del ds_gpm
            gc.collect()

    # Process and save each part
    process_and_save_part(files_part1, 1)
    process_and_save_part(files_part2, 2)
    process_and_save_part(files_part3, 3)

    # Additional garbage collection after processing all parts for the year
    gc.collect()

# Example usage
# folder = '/Users/aprida/Documents/Consulting/Public_sector/IDF_Togo/data/hydrography/gpm'  # Define your folder path
folder_year_files = f'{folder}/agg_files_per_year'
os.makedirs(folder_year_files, exist_ok=True)

# Set start and end year
year_start = 2013
year_end = 2013

# Define chunk size
chunk_size = 10000  # Use as a simple integer to chunk only the 'time' dimension

# Path to spatial reference file
spatial_ref_path = '/Users/aprida/.hydromt_data/artifact_data/v0.0.8/era5_hourly.nc'

for year in range(year_start, year_end + 1):
    process_and_save_files(year, folder, folder_year_files, chunk_size, spatial_ref_path)
    print(f"Completed processing for year {year}")
    gc.collect()  # Additional garbage collection after processing each year
```
