# Read raster layers from GeoNode (via WCS)

## Example 1: Digital Elevation Model of the Fiherenana River basin

```python
### Import Packages ###
from owslib.wcs import WebCoverageService
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import rioxarray as rxr

### Connect to Web Coverage Service (WCS) ###
wcs_url = "https://www.resiliencemada.gov.mg/gs/ows"
wcs = WebCoverageService(url=wcs_url, version="1.0.0")

### List available coverages ###
print("Available raster layers:")
for coverage in wcs.contents:
    print(coverage)

### Specify the layer name ###
layer_name = "geonode:fiherenana_elevtn"  # Replace with the correct name if different

### Get the bounding box and CRS ###
bbox = wcs[layer_name].boundingBoxWGS84  # Fetch the bounding box from metadata
crs = "EPSG:4326"                        # Use the CRS from the raster metadata

### Get the raster data ###
coverage_response = wcs.getCoverage(
    identifier=layer_name,
    format="image/tiff",
    bbox=bbox,
    crs=crs,
    width=500,  # Resolution or size of the raster
    height=500
)

### Save the raster file ###
fn = "/Users/aprida/Desktop/fiherenana_elevtn.tif"
with open(fn, "wb") as f:
    f.write(coverage_response.read())

### Read and Plot the Raster ###

# Read raster
raster = rxr.open_rasterio(fn)
masked_raster = raster.where(raster >= 0)

# Plot raster
fig, ax = plt.subplots(figsize=(15,10))
masked_raster.plot(cmap=plt.cm.terrain, cbar_kwargs={"label": "Elevation (m+MSL)"}, alpha=0.7, zorder=2, vmin=0, vmax=np.quantile(raster, 0.98))
gdf_aoi.plot(ax=ax, color='none', edgecolor='red', linewidth=5)
ax.set_title(label="Digital Elevation Model - Fiherenana River Basin")
ax.set_xlabel('Longitude (degrees)')
ax.set_ylabel('Latitude (degrees)')
cx.add_basemap(ax=ax, crs=4326, source=cx.providers.Esri.WorldImagery, attribution=False)

plt.show()
```

## Example 2: Locust Susceptibility map of Madagascar

```python
### Import Packages ###
from owslib.wcs import WebCoverageService
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import rioxarray as rxr
import matplotlib.colors as mcolors
import contextily as cx

### Connect to Web Coverage Service (WCS) ###
wcs_url = "https://www.resiliencemada.gov.mg/gs/ows"
wcs = WebCoverageService(url=wcs_url, version="1.0.0")

### Specify the layer name ###
layer_name = "geonode:Locust_Invasion_suitability_2050_SSP5_85_grma_madagascar_1"  # Replace with the correct name if different

### Get the bounding box and CRS ###
bbox = wcs[layer_name].boundingBoxWGS84  # Fetch the bounding box from metadata
crs = "EPSG:4326"                        # Use the CRS from the raster metadata

### Get the raster data ###
coverage_response = wcs.getCoverage(
    identifier=layer_name,
    format="image/tiff",
    bbox=bbox,
    crs=crs,
    width=500,  # Resolution or size of the raster
    height=500
)

### Save the raster file ###
fn = "/Users/aprida/Desktop/locust_suitability.tif"
with open(fn, "wb") as f:
    f.write(coverage_response.read())

### Read and Plot the Raster ###

# Read raster
raster = rxr.open_rasterio(fn)
masked_raster = raster.where(raster >= 0)

# Define a discrete colormap for values 0 to 3
cmap = mcolors.ListedColormap(["blue", "green", "yellow", "red"])  # Adjust colors as needed
norm = mcolors.BoundaryNorm(boundaries=[0, 1, 2, 3, 4], ncolors=4)

# Plot the raster with the custom colormap
fig, ax = plt.subplots(figsize=(8, 13))
masked_raster.plot(
    cmap=cmap,
    norm=norm,
    cbar_kwargs={
        "label": "Land Cover Classes",
        "ticks": [0.5, 1.5, 2.5, 3.5],  # Center the ticks within class ranges
        "format": "%d"  # Format tick labels as integers
    },
    alpha=0.7,
    zorder=2,
    ax=ax
)

# Add title and labels
ax.set_title(label="Locust Susceptibility")
ax.set_xlabel('Longitude (degrees)')
ax.set_ylabel('Latitude (degrees)')

# Add basemap
cx.add_basemap(ax=ax, crs=4326, source=cx.providers.Esri.WorldImagery, attribution=False)

# Show the plot
plt.show()
```

# Modify Metadata

Metadata reference: https://www.resiliencemada.gov.mg/api/v2/

## Get resources

It includes all files within GeoNode:

- Datasets
- Documents
- GeoApps
- Maps

There are different types of Metadata to add to each resource in GeoNode:

**1. Basic Metadata (mandatory)**:

- Title
- Abstract
- Date type
- Date
- Category (to be selected from predefined list of Categories)
- Group (to be selected from predefined list of Groups)
- Keywords (to be selected from predefined list of Keywords)

**2. Location and Licenses (mandatory)**:

- Language
- License (to be selected from predefined list of Licenses)
- Attribution
- Regions (to be selected from predefined list of Regions)
- Data quality statement
- Restrictions
- Other constraints


```python
import requests
import pandas as pd
import getpass

# Base URL for the categories endpoint

LOGIN_NAME = 'alvaro.prida@axaclimate.com'
LOGIN_PASSWORD = getpass.getpass("Enter your password: ")
auth = (LOGIN_NAME, LOGIN_PASSWORD)
```

```python
resources = ['datasets'] # ['documents', 'maps', 'geoapps']

df = pd.DataFrame()

for i in resources:
    
    BASE_URL = f"https://www.resiliencemada.gov.mg/api/v2/{i}/"

    # Initialize an empty list to store all resources
    all_resources = []
    
    # Pagination loop
    next_url = BASE_URL  # Start with the base URL
    while next_url:
        response = requests.get(next_url, auth=auth)
        if response.status_code == 200:
            data = response.json()
            # Add resources from the current page
            all_resources.extend(data[i])
            # Update the next_url for the next page
            next_url = data['links']['next']
        else:
            print(f"Error: {response.status_code}, {response.text}")
            break
    
    # Convert the list of resources to a pandas DataFrame
    df_tmp = pd.DataFrame(all_resources)
    df = pd.concat([df, df_tmp], ignore_index=True)

df = df[['name', 'raw_abstract', 'keywords', 'category', 'group']]

# Save to file
fn = '/Users/aprida/Desktop/datasets_250131.xlsx'
df.to_excel(fn, index=False, sheet_name='Datasets')

# Display the DataFrame
df
```
